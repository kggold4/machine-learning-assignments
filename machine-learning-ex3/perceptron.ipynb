> Perceptron

```python
import numpy as np
import matplotlib.pyplot as plt
from util_functions import create_data_from_file, plot_labels, mse, accuracy
```

```python
points, _ = create_data_from_file(file_name='two_circle.txt')

X = np.array([[p.x, p.y] for p in points])
y = np.array([p.label for p in points])

# plot the train data with the real labels of dots (before predicting)
plot_labels(X, y)
```

```python
class Perceptron(object):
    def __init__(self):
        self.errors_ = None
        self.learning_rate = 0.1
        self.weights = None
        self.bias = 0.1
        self.n_iter = 20
    def net_input(self, X):
        z = np.dot(X, self.weights[:]) + self.bias
        return z

    def predict(self, X):
        return np.array(np.where(self.net_input(X) >= 0, 1, -1))

    def fit(self, X, y):
        rgen = np.random.RandomState(1)
        self.weights = rgen.normal(loc=0.0, scale=0.01, size=1+X.shape[1])
        self.bias=self.weights[0]
        self.weights=self.weights[1:]
        self.errors_ = []
        self.accuracies_ = []
        for _ in range(self.n_iter):
            output = self.predict(X)
            errors = y - output
            self.weights[:] += self.learning_rate * X.T.dot(errors)

            # weights update
            self.bias += self.learning_rate * errors
            
            # calculation of cost and accuracy functions
            cost = mse(y, output)
            acc = accuracy(y, output)

            # cost and accuracy storage
            self.errors_.append(cost)
            self.accuracies_.append(acc)
        return self
```

```python
per_model = Perceptron()
per_model.fit(X ,y)
```

```python
plt.plot(per_model.errors_)
plt.title("Visualisation of errors")
plt.xlabel('Epochs')
plt.ylabel('Errors')
plt.show()
```

```python
plt.plot(per_model.accuracies_)
plt.title("Visualisation of accuracies")
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```

```python
predicted = per_model.predict(X)
print(f"Final vector of predicted labels are:\n{predicted}")
plot_labels(X, predicted)
```

```python
print(f"Train data accuracy: {accuracy(y, predicted) * 100} %")
```
